```python
import customtkinter as ctk

# Configuração inicial
ctk.set_appearance_mode("light")
ctk.set_default_color_theme("blue")

# Criar janela principal
login_window = ctk.CTk()
login_window.geometry("700x400")
login_window.title("Login")
login_window.resizable(width=False, height=False)

# Configuração da grade
num_linhas = 5  # Defina quantas linhas você quer no grid
for i in range(num_linhas):
    login_window.grid_rowconfigure(i, weight=1)

for i in range(3):  # Número de colunas (opcional)
    login_window.grid_columnconfigure(i, weight=1)

# Criar frame para ocupar todas as linhas
frame_logo = ctk.CTkFrame(login_window, fg_color="blue")
frame_logo.grid(row=0, column=0, rowspan=num_linhas, padx=10, pady=10, sticky="nswe")  # Usa num_linhas no rowspan

# Exemplo de outro frame ao lado
frame_form = ctk.CTkFrame(login_window, fg_color="gray")
frame_form.grid(row=0, column=1, columnspan=2, padx=10, pady=10, sticky="nswe")  # Ocupa as colunas 1 e 2

# Iniciar a interface
login_window.mainloop()

```

```python
import customtkinter as ctk
from PIL import Image

# Configuração inicial
ctk.set_appearance_mode("light")
ctk.set_default_color_theme("blue")

# Criar janela principal
login_window = ctk.CTk()
login_window.geometry("600x400")
login_window.title("Login")
login_window.resizable(width=False, height=False)

# Configuração da grade
num_linhas = 11
for i in range(num_linhas):
    login_window.grid_rowconfigure(i, weight=1)

for i in range(5):
    login_window.grid_columnconfigure(i, weight=1)

# Criar frame para ocupar todas as linhas
frame_logo = ctk.CTkFrame(login_window)
frame_logo.grid(row=0, column=0, rowspan=num_linhas, padx=0, pady=0, sticky="nswe")

# Configurar o grid dentro do frame_logo
for i in range(num_linhas):
    frame_logo.grid_rowconfigure(i, weight=1)
frame_logo.grid_columnconfigure(0, weight=1)

# Verificar e carregar a imagem
img = Image.open(r"C:\Users\miria\OneDrive\Área de Trabalho\mbk-email-mei-projeto\interface_grafica\logo_sem_fundo.jpg")  # Substitua pelo caminho correto
login_window.logo_mbk = ctk.CTkImage(
    light_image=img,
    dark_image=img,
    size=(100, 90)
)

# Adicionar um CTkLabel com a imagem
logo_label = ctk.CTkLabel(frame_logo, image=login_window.logo_mbk, text="")
logo_label.grid(row=3, column=0, padx=10, pady=10, sticky="nswe", rowspan=3)


# Título da aplicação com fonte maior e centralizado
title = ctk.CTkLabel(
    login_window,
    text="Painel de Controle de E-mails",
    font=("Segoe UI", 24, "bold")
)
title.grid(row=1, column=2, padx=10, pady=10, sticky="nswe")

# Iniciar a interface
login_window.mainloop()
```
